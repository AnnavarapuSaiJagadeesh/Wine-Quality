# Preprocessing

```python
X = df.drop("quality", axis = 1)
y = df["quality"]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5, stratify=y)

print(y_train.value_counts())
y_test.value_counts()
```

# SVM one class classifier

```python
# outlier detection algorithm is used
from sklearn.svm import OneClassSVM

model = OneClassSVM(kernel="rbf",nu = 0.0625) # why 0.0625 ? because 16/256 = 0.625
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
predict = np.where((predict == -1), 0, predict)
# print(np.array(y_test))
# print(predict)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(accuracy_score(predict, y_test))
print(classification_report(predict, y_test))
```

# Feature Selection followed by SVM One class classifier

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif

for no_of_features in range(1,12):
  print("Using top ",no_of_features, " features")

  # To selelect top "no_of_features"
  selector = SelectKBest(score_func = mutual_info_classif,k = no_of_features)
  X_train_selected = selector.fit_transform(X_train, y_train)
  X_test_selected = selector.transform(X_test)

  # Top features
  selected_feature_names = df.drop("quality",axis = 1).columns[selector.get_support()].tolist()
  print("features used are : ",selected_feature_names)


  # Training the model
  model = OneClassSVM(kernel="rbf",nu = 0.0625)
  model.fit(X_train_selected, y_train)
  predict = model.predict(X_test_selected)
  predict = np.where((predict == -1), 0, predict)
  # print(accuracy_score(predict, y_test))
  print(classification_report(predict, y_test))
  print("-------------------------------------------------------------------")
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predict = model.predict(X_test)

print(classification_report(predict, y_test))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 100, class_weight = "balanced", random_state = 5)
classifier.fit(X_train, y_train)
predict = classifier.predict(X_test)
# print(predict)
# print(np.array(y_test))

print(classification_report(predict, y_test))
```
